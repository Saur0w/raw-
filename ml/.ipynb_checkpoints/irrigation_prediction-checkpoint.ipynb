```python
import pandas as pd
import numpy as np
import joblib
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("irrigation_prediction.csv")
df.head()
```

```python
df = df[["Temperature_C", "Humidity", "Soil_Moisture", "Rainfall_mm", "Sunlight_Hours", "Wind_Speed_kmh", "Soil_pH", "Irrigation_Need",]]
df.head()
```

```python
print(df.info())
print(df.isnull().sum())
```

```python
X = df.drop("Irrigation_Need", axis=1)
y = df["Irrigation_Need"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
model = RandomForestClassifier(n_estimators=100)
model.fit(X_train,y_train)
```

```python
result = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, result))
print("\nClassification Report:\n", classification_report(y_test, result))
```

```python
joblib.dump(model, "irrigation_model.pkl")
```

```python
sample_input = [[
    40,   # Temperature
    50,   # Humidity
    40,   # Soil Moisture
    5,    # Rainfall
    2,    # Sunlight
    10,   # Wind Speed
    6.5   # Soil pH
]]

prediction = model.predict(sample_input)[0]
print("Prediction:", prediction)
```
